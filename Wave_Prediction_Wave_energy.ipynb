```python
def visualize_monthly_trends(df, column, title):
    """
    Visualizes the monthly trend of a selected column.

    Parameters:
    - df: Pandas DataFrame containing the dataset.
    - column: Column name (string) to visualize.
    - title: Custom title for the plot.
    """
    # Convert time column to datetime format (if not already)
    df["time"] = pd.to_datetime(df["time"], errors="coerce")
    
    # Extract year and month
    df["Year-Month"] = df["time"].dt.to_period("M")
    
    # Group by month and calculate mean
    monthly_data = df.groupby("Year-Month")[column].mean().reset_index()
    
    # Convert back to datetime for plotting
    monthly_data["Year-Month"] = monthly_data["Year-Month"].astype(str)

    # Plot the trend
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=monthly_data["Year-Month"], y=monthly_data[column], marker="o")
    plt.title(f"Monthly Trend of {title}", fontsize=14)
    plt.xlabel("Month", fontsize=12)
    plt.ylabel(title, fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
# Convert relevant columns to numeric types
wave_df["wave_ht_sig"] = pd.to_numeric(wave_df["wave_ht_sig"], errors='coerce')
wave_df["wind_spd_avg"] = pd.to_numeric(wave_df["wind_spd_avg"], errors='coerce')
wave_df["wave_ht_max"] = pd.to_numeric(wave_df["wave_ht_max"], errors='coerce')
# Now we can visualize the monthly trends again
visualize_monthly_trends(wave_df, "wave_ht_max", "Maximum Wave Height")
visualize_monthly_trends(wave_df, "wave_ht_sig", "Significant Wave Height")
visualize_monthly_trends(wave_df, "wind_spd_avg", "Average Wind Speed")
```

```python
time = wave_df['time']
max_wave_height = wave_df['wave_ht_max']
plt.figure(figsize=(20, 10))
plt.plot(time, max_wave_height)
# x_ticks = pd(np.arange(0, len(time), 1000), rotation=45)
plt.xticks(np.arange(0, len(time), 1440), rotation=45)
# plt.yticks(np.arange(0, max(max_wave_height)+5, 1))
# plt.yticks(y_ticks)
plt.xlabel('Time')
plt.ylabel('Wave Height (m)')
plt.title('Wave Height Over Time')
plt.show()
```

```python
for col in wave_df.columns:
    if col not in ["time", "station_name"]:
        wave_df[col] = pd.to_numeric(wave_df[col], errors='coerce')

# Select only numeric columns for correlation matrix
numeric_cols = wave_df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_cols.corr()

# Display correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Plot correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
numeric_cols.hist(figsize=(15, 10), bins=30, edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=wave_df["wind_spd_avg"], y=wave_df["wave_ht_sig"], alpha=0.5)
plt.xlabel("Average Wind Speed (m/s)")
plt.ylabel("Significant Wave Height (m)")
plt.title("Wind Speed vs. Wave Height")
plt.show()
```

```python
# Boxplots for numerical features
plt.figure(figsize=(15, 8))
sns.boxplot(data=numeric_cols)
plt.xticks(rotation=90)
plt.title("Boxplot of Numerical Features")
plt.show()
```
