# **Exploratory Data Analysis**

```python
time = wave_df['time']
max_wave_height = wave_df['wave_ht_max']
plt.figure(figsize=(20, 10))
plt.plot(time, max_wave_height)
# x_ticks = pd(np.arange(0, len(time), 1000), rotation=45)
plt.xticks(np.arange(0, len(time), 1440), rotation=45)
# plt.yticks(np.arange(0, max(max_wave_height)+5, 1))
# plt.yticks(y_ticks)
plt.xlabel('Time')
plt.ylabel('Wave Height (m)')
plt.title('Wave Height Over Time')
plt.show()
```
